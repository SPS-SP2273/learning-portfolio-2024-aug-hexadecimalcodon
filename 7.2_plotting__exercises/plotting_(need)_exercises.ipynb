# Exercise 1 (Fill in the blanks)

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color='darkred', linestyle='dashed',
             marker='o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label='Y values with errors')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')
plt.show()
```

# Exercise 2 (A simple mathematical plot)  



```python
import numpy as np
```

```python
x = np.array([0,10,20,30,40,50])
e = np.e
y = (e**(-x/10))*(np.sin(x))
```

```python
x1 = np.linspace(0, 50, 1000)
y1 = (e**(-x1/10))*(np.sin(x1))   # when using np.linspace, x1 = np.linspace (# of plot points) || y1 = y eqn. with x1 instead of x
plt.plot(x1, y1, label='Y values',
         color='blue')
plt.xlabel('x', fontsize = 15, color = 'darkorange')
plt.ylabel('y', fontsize = 15, color = 'darkorange')
plt.title('y =$e^{-x/10}sin(x)$', fontsize = 20)
plt.grid(alpha=.25)
plt.ylim([-1,1])
plt.show()
```
